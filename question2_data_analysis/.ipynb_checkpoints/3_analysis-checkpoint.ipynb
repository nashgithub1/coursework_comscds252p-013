```python
import pandas as pd
```

```python
df = pd.read_csv('books_cleaned.csv')
```

```python
#Central Tendency
mean_price = df['Price'].mean()
median_price = df['Price'].median()
mode_price = df['Price'].mode()[0]

print("mean price is : ",mean_price)
print("median price is : ",median_price)
print("mode price is: ", mode_price)
```

```python
#Dipersion
std_price = df['Price'].std()
price_range = df['Price'].max() - df['Price'].min()

print("Standard price is : ",std_price)
print("price range is : ",price_range)
```

```python
#Top 5 Categories by Average Price
top_categories = df.groupby('Category')['Price'].mean().sort_values(ascending=False).head(5)
print(top_categories)
```

```python
#Rating Distribution
rating_counts = df['Rating'].value_counts().sort_index()
print(rating_counts)
```

```python
#Statistical Analysis
#1. Inferencial Statistics
```

```python
#Outlier Detection
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['Price'] < Q1 - 1.5*IQR) | (df['Price'] > Q3 + 1.5*IQR)]

print(outliers)
```

```python
#Correlation
correlation = df['Price'].corr(df['Rating'])
print(correlation)
```

```python
#Hypothesis Testing (Fiction v Non Fiction
from scipy.stats import ttest_ind

fiction = df[df['Category'].str.contains('Fiction', case=False, na=False)]['Price']
nonfiction = df[~df['Category'].str.contains('Fiction', case=False, na=False)]['Price']

t_stat, p_val = ttest_ind(fiction, nonfiction)

print("T-stat:", t_stat)
print("P-value:", p_val)
```
