```python
#Histogram
import pandas as pd
df = pd.read_csv('books_cleaned.csv')
import matplotlib.pyplot as plt

plt.hist(df['Price'], bins=20)
plt.axvline(df['Price'].mean())
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
#Box Plot
df.boxplot(column='Price', by='Category')
plt.xticks(rotation=90)
plt.title("Price by Category")
plt.show()
```

```python
#Scatter Plot
#trying seaborn for this only

import seaborn as sns
plt.figure(figsize=(8,5))
sns.regplot(data=df, x='Rating', y='Price', scatter_kws={'alpha':0.6})
plt.title("Price vs Rating with Trend Line")
plt.xlabel("Rating")
plt.ylabel("Price")
plt.show()
```

```python
#Bar Chart
df.groupby('Category')['Rating'].mean().sort_values(ascending=False).head(8).plot(kind='bar')
plt.title("Avg Rating by Category")
plt.show()
```
