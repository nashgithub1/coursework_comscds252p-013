```python
import pandas as pd

df = pd.read_csv('books_uncleaned.csv')
```

```python
# Clean column names
df.columns = df.columns.str.strip()
```

```python
# Clean price
# Got error when cleaning only Â£ (some latin type encoding issue)
# so cleaned every special character
df['Price'] = df['Price'].str.replace(r'[^0-9.]', '', regex=True).astype(float)
```

```python
#check whether everything is ok
print(df.head())
print(df.tail())
```

```python
#Got 100 records
```

```python
# Rating conversion
rating_map = {"One":1, "Two":2, "Three":3, "Four":4, "Five":5}
df['Rating'] = df['Rating'].map(rating_map)
```

```python
# Missing values
print(df.isnull().sum())
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Price category
def price_category(price):
    if price < 20:
        return "Budget"
    elif price <= 40:
        return "Mid-range"
    else:
        return "Premium"

df['price_category'] = df['Price'].apply(price_category)
```

```python
# Stock availability
df['in_stock'] = df['Availability'].apply(lambda x: True if "In stock" in x else False)

```

```python
# Save cleaned data
df.to_csv('books_cleaned.csv', index=False)
```
